```python
#EDA variables categoricas
def graficos_eda_categorias(cat):

    #Calcular el numero de filas que necesitamos
    from math import ceil
    filas = ceil (cat.shape[1]/2)

    #Definimos el grafico
    f, ax = plt.subplots(nrows= filas, ncols=2, figsize= (16, filas * 16))

    #Aplanamos el grafico
    ax = ax.flat

    #Creamos el bucle
    for cada, variable in enumerate(cat):
      cat[variable].value_counts().plot.barh(ax = ax[cada])
      ax[cada].set_title(variable, fontsize =12, fontweight = 'bold')
      ax[cada].tick_params(labelsize = 12)
```

```python
def estadisticos_cont(num):

  #Calcular describe en las columnas
  etds = num.describe().T

  #anadimos la mediana
  etds['median'] = num.median()

  #Reordenamos para que la mediana esta al lado de la media
  etds = etds.iloc[:,[0,1,8,2,3,4,5,6,7]]

  #devolver o retornar la función
  return(etds)
```

¿Cual es el impacto económico de este problema? Según el estudio "Cost ol Tumover- del Center lor American Progress:

El coste de la fuga de los empleados que ganan menos de 30000 es del 16,1% de su salario

El coste de La fuga de los empleados que ganan entre 30000 - 50000 es del 19.7% de su salario

El coste de la fuga de los empleados que ganan entre 50000 - 75000 es del 20.4% de su salario

El Coste de la fuga de los empleados que ganan más de 75000 es del 21% de su salario


```python
#Creamos una variable salario_año del empleado
df['salario_ano'] = df.salario_mes.transform(lambda x: x*12)
df[['salario_mes','salario_ano']]
```

```python
#calcular el inpacto economico de cada empleado si deja la empresa

conds = [(df['salario_ano'] < 30000),(df['salario_ano'] >= 30000) & (df['salario_ano'] < 50000),(df['salario_ano'] >= 50000) & (df['salario_ano'] < 75000),(df['salario_ano'] >= 75000)]


resuls = [df.salario_ano * 0.161, df.salario_ano * 0.197, df.salario_ano * 0.204, df.salario_ano * 0.21]


#Aplicación de las Condiciones

df['impacto_abandono'] = np.select(conds,resuls , default= 000)

df
```

***¿Cuanto Nos a costado el problema el ultimo año?***

```python
costo_total = df.loc[df.abandono == 1].impacto_abandono.sum()
print('El impacto que se tuvo en la empresa del ultimo año fue:' , costo_total)
```

*¿Cuanto Nos cuenta los empleados que no estan motivados? *

```python
# cuantos Empleados se fueron
total_ps = len(df.loc[df.puesto == 'Sales Representative'])

abandono_ps = len(df.loc[(df.puesto == 'Sales Representative') & (df.abandono == 1)])
porc_ps = abandono_ps / total_ps

print('El % de abandono en la empresa fue de:' , porc_ps * 100)
```

***Hipotecis***

```python
# cuantos Empleados se me pueden ir actualmente
total_ac = len(df.loc[(df.puesto == 'Sales Representative') & (df.abandono == 0)])

se_retiran = int(total_ac * porc_ps)

print('Sales Representative que se retiran de la compañia este año es de:' , se_retiran ,'personas')
```

***Hipoteisis de 30% para saber cuanto dienero puedo suponer para la compañia?***

```python
ret = int(se_retiran * 0.3)

ahorro = df.loc[(df.puesto == 'Sales Representative') & (df.abandono == 0), 'impacto_abandono'].sum() * porc_ps * 0.3

print(f'podemos retener {ret} representantes de ventas y ello supondria en dinero: $ {int(ahorro)} dolares')
```

***PREPARACIÓN DE DATOS DEL MODELO ***

Transformar todos las variables categoricas a numericas

```python
from sklearn.preprocessing import OneHotEncoder

#Catecoricas
cat = df_ml.select_dtypes('O') # se le pasa la O mayuscula para que haga referencia a las variables que son Texto y las traiga todas

#Instancimos
ohe = OneHotEncoder(sparse = False)

#Entrenamos
ohe.fit(cat)

#Aplicacamos la transformación
cat_ohe = ohe.transform(cat)

#Se Evaluan las variables Categoricas
cat_ohe = pd.DataFrame(cat_ohe, columns= ohe.get_feature_names_out(input_features= cat.columns)).reset_index(drop = True)
```

```python
cat_ohe
```

```python
num = df.select_dtypes('number').reset_index(drop = True)

#Juntamos las variables
df_ml = pd.concat([cat_ohe,num], axis = 1)
```

```python
df_ml
```

```python
#Separación de variables predictoras
df_ml = df_ml.dropna()
x = df_ml.drop(columns='abandono')
y = df_ml['abandono']

#Separación de entrenamiento y test
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.3)

from sklearn.tree import DecisionTreeClassifier

#Instanciar
c = DecisionTreeClassifier(max_depth=4)

#Entrenar
c.fit(train_x,train_y)
```

```python
#Predicción
pred = c.predict_proba(test_x)[:,1]

pred[:20] #El numero de personas que quiero validar este lo escoge al azar
```

```python
#Evaluar modelo
from sklearn.metrics import roc_auc_score

eva = roc_auc_score(test_y, pred)
print(f'El % de evalución de nuestro modelo es: {eva*100}')
```

```python
#Driagrama de arbol
from sklearn.tree import plot_tree

plt.figure(figsize= (50,50))

plot_tree(c,feature_names = test_x.columns,impurity=False,node_ids=True,proportion=True,rounded=True,precision=2);
```

```python
#Importancia de estas variables
pd.Series(c.feature_importances_, index = test_x.columns).sort_values(ascending = False).plot(kind = 'bar', figsize = (30,20))
```

```python
df_ml['Scoring_abandono'] = c.predict_proba(df_ml.drop(columns='abandono'))[:, 1]
```
